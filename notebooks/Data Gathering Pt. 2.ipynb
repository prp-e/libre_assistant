## Storing the chunks

```python
import math
chunk_size = 6144
```

```python
no_chunks = math.ceil(len(big_str)/chunk_size) 
```

```python
no_chunks
```

```python
dataset = []
```

```python
for i in range(0, len(big_str), chunk_size):
    chunk = big_str[i:i+chunk_size]
    dataset.append(chunk)
```

## Building the dataset

```python
frame = []

for data in dataset:
    string = data.split('.')
    prompt = string[0]
    del string[0]
    completion = '.'.join(string)
    
    frame_chunk = [prompt, completion]
    frame.append(frame_chunk)
```

```python
len(frame)
```

```python
cols = ['prompt', 'completion']
```

```python
new_df = pd.DataFrame(frame, columns=cols)
```
